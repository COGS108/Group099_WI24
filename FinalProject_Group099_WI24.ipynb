# Abstract

In this study, we aimed to investigate the relationship between public transportation usage and the frequency of car accidents in a city. In addition to that, we also wanted to investigate the relationship between the severity of car accidents and public transportation usage. Our research question focuses on whether or not higher usage of public transportation correlates with a decrease in the number of car accidents and a reduction in their severity, i.e. a negative linear relationship. Our alternate hypothesis proposes that increased rates of public transportation usage could contribute to safer road environments due to factors such as professional public transportation drivers better adhering to safety protocols, reduced traffic congestion, and fewer reckless drivers on the road. The null hypothesis states that there were no correlation between public transportation usage and number of car accidents and their severity.

To conduct our study, we used datasets three datasets to explore our research question. One of the factors being that public transportation involves professional drivers who go through rigorous training to operate the mode of transportation that they are in charge of and may adhere more strictly to traffic regulations and safety protocols which can reduce the likelihood of accidents. ## Dataset #3 - City and Town Population Totals: 2010-2019

````python
population_df = pd.read_excel('SUB-IP-EST2019-ANNRES-06.xlsx')

population_df.head()
````

````python
# The code does not have proper column markers, and has excess information. We will fix that with the following:

# Fix column values
col_values = population_df.iloc[2]
col_values.iloc[0] = population_df.iloc[1].iloc[0]
population_df.columns = col_values

# Simplify rows and columns
population_df = population_df[3:]
population_df = population_df.reset_index(drop=True).get(['Geographic Area', 2017.0]).rename(columns={'Geographic Area': 'City', 2017.0:'population_2017'})
population_df['City'] = population_df['City'].str.title().str.replace(', California', '').str.replace(' City', '')
population_df.dropna(inplace=True)

    
population_df.head()
````

### __Part 2.1: Exploring relationships with "Accident Rate"__
In this section, we will run a linear regression model to see if there is a relationship between public transportation usage and the amount of accidents that occur in a city. We begin by removing outliers and examining if there is a trend in our data. We chose to study "Accident Rate" and "Percent Using Transit" applied with a log10 transformation to shift the values away from being centered near zero when put on the log scale.




````python
# create a subset of clean_df for column of interest for linear regression model
sub_col_acc_transit = ['accident_rate','transit_usage']
sub_accident_rate = clean_df[sub_col_acc_transit]

# Apply a log-10 transformation for less skewed dataset

sub_accident_rate['transit_log10'] = np.log10(sub_accident_rate['transit_usage']+0.1)
sub_accident_rate.head()
````

````python
sub_col_2 = ['accident_rate','transit_usage']
sub_accident_rate_2 = clean_df[sub_col_2]

# Apply a log-10 transformation for less skewed dataset

sub_accident_rate_2['transit_log10'] = np.log10(sub_accident_rate_2['transit_usage']+0.1)

# filter out the outliers from 'accident_rate'
x = 2
mean = sub_accident_rate_2['accident_rate'].mean()
std = sub_accident_rate_2['accident_rate'].std()
filtered_accidents  = sub_accident_rate_2[(sub_accident_rate_2['accident_rate'] > (mean - x * std)) & (sub_accident_rate_2['accident_rate']< (mean + x * std))]
filtered_accidents

filtered_accidents.head()
````

<ins>Linear Regression of Accident Rate and Percent Using Transit:</ins>

````python
# Create a figure and axes objects for subplots
sns.regplot(x='transit_log10', y='accident_rate', data=filtered_accidents, fit_reg=True, color='#c74b2c')
plt.xlabel('transit_log10')
plt.ylabel('accident rate')
plt.title('Relationship between Accident Rate and Transit Usage')

# Show the plots
plt.show()
````

Observations:
1. The best fit line slopes upwards, which implies a positive linear relationship between public transportation use rates and the rate at which accidents occur. This goes against our initial hypothesis that transit usage has a negative effect on accident occurrence.
2. We observe that the points are scattered. There may be little correlation between the two variables.

### __Part 2.2: Finding evidence for correlation with Accident Rate__

<ins>OLS Model of Accident Rate and Percent Using Transit:</ins>

````python
outcome,predictors= patsy.dmatrices('accident_rate ~ transit_log10',filtered_accidents)
model = sm.OLS(outcome,predictors)
res_log = model.fit()
print(res_log.summary())
````

Through our summary, we can observe the following:
1. The p-value is 0.035, which is less than our significance level a = 0.05. This is statistically significant, and allows us to reject our null hypothesis that public transportation usage does not affect accident rates.
2. R-sqared is 0.079, which is much closer to 0 than 1. This means that our plot is scattered, and the variance cannot be explained by transit usage alone.

From running an OLS model, we found that our p-value was 0.035, which is less than our significance level of 0.05.   This is statistically significant so we can reject the null hypothesis that public transportation usage does not affect the accident rate. However, it should be noted that R-squared is 0.079, which means very little of the variance in the accident rates can be explained by transit usage alone. Because our regression line was upwards sloping, which implies an increase in transit usage leads to an increase in accident rates, we can say that our alternate hypothesis was not supported by our research.  
<br>
The second set of analyses explored our second research question: does increased public transportation usage lessen the severity of accidents? From the linear regression model, we found that we cannot reject our null hypothesis for Accident Severity since the p-value is 0.071 which is bigger than 0.05. This is statistically insignificant so we cannot reject the null hypothesis of public transportation affecting the accident severity. Additionally, our  R-squared value is 0.040 which is closer to 0, meaning it overall has a poor fit.  
<br>
Ultimately, our research concluded that there is no significant change in accident rate or severity as caused by increased public transportation usage. There are many other confounding variables that we failed to find initially, which may allow us to better understand the effect of public transportation usage on accident occurrence.

# Discussion and Conclusion

There was, however, research that hinted towards many benefits of public transportation, such as lessening traffic congestion. This led us to believe that a higher use of public transportation would create a “domino effect”: public transportation would indicate fewer vehicles on the road, which could mean fewer reckless drivers, which could mean fewer car crashes. We also believe that public transportation drivers are better equipped to handle accidents, as they have more training so they can better react to dangerous situations, leading us to wonder if higher public transportation usage can also lessen the average severity of car crashes. Through our background research, we hypothesized that public transportation could serve as a remedy to those issues since it gives an option for unfit drivers to leave the wheel and let professionals, who go through rigorous training to get the appropriate license, to take to the roads instead. Thus, we predicted that a higher utilization of public transportation could reduce the number of accidents on the road as well as the severity of the accident.

While both of the linear regression models we created for comparing transit usage to accident rates and severity had low p-values, at 0.035 and 0.071 respectively, there was a high level of variance and positive linear relationship which went against our alternative hypothesis that a higher rate of public transportation usage would lead to lower accident rates.
