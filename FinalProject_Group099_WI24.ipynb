## Dataset #3 - City and Town Population Totals: 2010-2019

````python
population_df = pd.read_excel('SUB-IP-EST2019-ANNRES-06.xlsx')

population_df.head()
````

````python
# The code does not have proper column markers, and has excess information. We will fix that with the following:

# Fix column values
col_values = population_df.iloc[2]
col_values[0] = population_df.iloc[1][0]
population_df.columns = col_values

# Simplify rows and columns
population_df = population_df[3:]
population_df = population_df.reset_index(drop=True).get(['Geographic Area', 2017.0]).rename(columns={'Geographic Area': 'City', 2017.0:'Population in 2017'})
population_df['City'] = population_df['City'].str.title().str.replace(', California', '').str.replace(' City', '')
population_df.dropna(inplace=True)

    
population_df.head()
````

````python
#We will merge this into the dataset 'accidents_vs_transit' to get the per-capita accident rate of each city.

clean_df = pd.merge(accidents_vs_transit, population_df, on='City')

clean_df = clean_df.assign(accident_rate = clean_df['num_accidents']/clean_df['Population in 2017'])

clean_df = clean_df.get(['City', 'num_accidents', 'Population in 2017', 'accident_rate', 'avg_severity', 'Percent Using Transit'])

# Look at some values in our cleaned dataframe
clean_df.head(10)
````

````python
# create a subset of clean_df for column of interest for linear regression model
sub_col_acc_transit = ['accident_rate','Percent Using Transit']
sub_accident_rate = clean_df[sub_col_acc_transit]

# Apply a log-10 transformation for less skewed dataset

sub_accident_rate['transit_log10'] = np.log10(sub_accident_rate['Percent Using Transit']+0.1)
sub_accident_rate.head()
````

````python
sub_col_2 = ['accident_rate','Percent Using Transit','Population in 2017']
sub_accident_rate_2 = clean_df[sub_col_2]

# Apply a log-10 transformation for less skewed dataset

sub_accident_rate_2['transit_log10'] = np.log10(sub_accident_rate_2['Percent Using Transit']+0.1)

# filter out the outliers from 'Population in 2017'
x = 2
mean = sub_accident_rate_2['Population in 2017'].mean()
std = sub_accident_rate_2['Population in 2017'].std()
filtered_pop  = sub_accident_rate_2[(sub_accident_rate_2['Population in 2017'] > (mean - x * std)) & (sub_accident_rate_2['Population in 2017']< (mean + x * std))]
filtered_pop

# Divide 'Population in 2017'by 1e8 to make it comparable in range of accident rate 
filtered_pop['pop_comp'] = filtered_pop['Population in 2017']/1e8
filtered_pop.head()
````

````python
fig, ax = plt.subplots(figsize=(14, 8))


sns.regplot(x='transit_log10', y='accident_rate', data=filtered_pop, fit_reg=True, ax=ax, label='Accident Rate')
sns.regplot(x='transit_log10', y='pop_comp', data=filtered_pop, fit_reg=True, ax=ax, label='Population in 2017')
ax.set(ylabel='accident rate / (Population / 1e8)', xlabel='transit_log10')


plt.title('Relationship between Accident Rate and Percent Using Transit')
ax.legend()
plt.show()
````

````python
outcome,predictors= patsy.dmatrices('accident_rate ~ transit_log10',sub_accident_rate)
model = sm.OLS(outcome,predictors)
res_log = model.fit()
print(res_log.summary())
````

````python
outcome,predictors= patsy.dmatrices('accident_rate ~ pop_comp',filtered_pop)
model = sm.OLS(outcome,predictors)
res_log = model.fit()
print(res_log.summary())
````

````python
potential_confound = 'accident_rate ~ transit_log10 + pop_comp'
outcome_2, predictors_2 = patsy.dmatrices(potential_confound, filtered_pop)

model_2 = sm.OLS(outcome_2, predictors_2)
res_2 = model_2.fit()
print(res_2.summary())
````
